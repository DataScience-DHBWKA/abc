```python
#Gruppierungen in 0.1 Schritten von 0.4 bis 1.0
def custom_rename(value):
    if 0.4 <= value < 0.5:
        return 'Group_0.4'
    elif 0.5 <= value < 0.6:
        return 'Group_0.5'
    elif 0.6 <= value < 0.7:
        return 'Group_0.6'
    elif 0.7 <= value < 0.8:
        return 'Group_0.7'
    elif 0.8 <= value < 0.9:
        return 'Group_0.8'
    elif 0.9 <= value < 1.0:
        return 'Group_0.9'
    else:
        return 'Other_Value'

#HDI Gruppieren
merged_data['HDI_neu'] = merged_data['hdi'].apply(custom_rename)
```
